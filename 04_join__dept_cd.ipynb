```python
# 필요 라이브러리 호출
import pandas as pd
import numpy as np
from numpy import NaN, NAN, nan
```

```python
# 회의록 df 호출
meeting_records = pd.read_csv("../First_project/dataes/03_21st_main_conf_total.csv", index_col = 0)
meeting_records
```

```python
# 의원명부 df 호출
members = pd.read_csv("../First_project/dataes/Congress_members_fin.csv")
members.head(5)
```

```python
# 울릉군 김병욱(9771285) - 35
# 서울동작을 이수진(9771284) - 192
# 비례대표 이수진(9771269) - 193
# 이상 3명은 동명이인이면서 발언기록이 없는 의원 3명임
# 제거할 df 생성
del_target = members.iloc[[35, 192, 193]].index
del_target
```

```python
# members_1 변수명에 제거할 해당의원 3명 제거
members_1 = members.drop(del_target)
members_1.head(1)
```

```python
# 조인에 사용할 컬럼을 추출하여 members_2 변수에 대입
members_2 = members_1[['dept_cd', 'con_name_kr']]
members_2
```

```python
# meeinng_record 변수에 호출한 df의 name컬럼 왼쪽에 'dept_cd' 컬럼 추가(컬럼 내 값은 NaN)
meeting_records.insert(3, 'dept_cd', NaN)
meeting_records.head()
```

```python
# meeting_records_1 변수에 'name'과 'dept_cd' 조인
meeting_records_1 = pd.merge(meeting_records,
                             members_2,
                             on = 'con_name_kr', how = 'left',
                             sort = False)
meeting_records_1.head()
```

```python
# dept_cd_x에 dept_cd_y의 값을 대입
meeting_records_1['dept_cd_x'] = meeting_records_1['dept_cd_y']
meeting_records_1.head()
```

```python
# dept_cd_y컬럼은 필요없으므로 삭제
tmp = meeting_records_1.drop('dept_cd_y', axis = 1)
tmp
```

```python
# 컬럼명 변경('dept_cd_x'->'dept_cd')
meeting_records_2 = tmp.rename(columns = {'dept_cd_x':'dept_cd'})
meeting_records_2.head()
```

```python
# 결측치 확인
print(meeting_records_2.count())
```

```python
# 결측치 제거
meeting_records_3 = meeting_records_2.dropna()
meeting_records_3
```

```python
# 컬럼별 데이터형 확인
meeting_records_3.info()
```

```python
# dept_cd 형변환(float -> int)
meeting_records_3 = meeting_records_3.astype({"dept_cd":'int'})
```

```python
# 인덱스 초기화
meeting_records_3.reset_index(drop = True, inplace = True)
meeting_records_3
```

```python
# 바꿔야할 칼럼명 확인
meeting_records_3.columns
```

```python
meeting_records_3.info()
```

```python
# date컬럼 datetime형으로 변환
meeting_records_3['date'] = meeting_records_3['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
```

```python
meeting_records_3.to_csv('../First_project/dataes/04_21st_main_conf_change_columns.csv')
```
