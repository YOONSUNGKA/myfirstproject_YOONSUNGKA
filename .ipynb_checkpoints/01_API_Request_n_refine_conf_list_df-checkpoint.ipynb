```python
# 라이브러리 호출
import requests
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np
```

```python
# 서비스키 인증 후 url 변수지정, params 딕셔너리 지정
# en_key는 사용하지 않음
# en_key = '4XpGdn1j7RgYNbyX74%2Bp8nry1Feq2ADLhP9Q0a78b8YPBfyoJ2f0FnnoNLzEH3DVwSFH0KLY9UTWD8Ofq%2Bha7g%3D%3D'
de_key = '4XpGdn1j7RgYNbyX74+p8nry1Feq2ADLhP9Q0a78b8YPBfyoJ2f0FnnoNLzEH3DVwSFH0KLY9UTWD8Ofq+ha7g=='
url = 'http://apis.data.go.kr/9710000/ProceedingInfoService/getAllConInfoList?'
# serviceKey(서비스키): de_key
# numOfRows(행개수): 10000
# PageNo(페이지번호): 1
# dae_num(국회대수): 21
# class_code(회의구분코드) : 1(본회의)
params = {'serviceKey':de_key,
          'numOfRows' : '10000',
          'pageNo' : '1',
          'dae_num' : '21',
          'class_code' : '1'}
```

```python
# 회의목록 긁어오기, 정렬 및 인덱싱, csv파일 저장
response = requests.get(url, params = params)
root = ET.fromstring(response.text)

print(response.content)
```

```python
conferNum = []  # (필수) 회의번호
commName = []   # (필수) 위원회
meeting1 = []   # (필수) 회기별
meeting2 = []   # (필수) 차수별
hwpLink = []    # (옵션) 회의록(hwp파일)
summary = []    # (옵션) 요약정보

for item in root.findall('.//item'):
    conferNum_tmp = item.find('conferNum').text
    commName_tmp = item.find('commName').text
    meeting1_tmp = item.find('meeting1').text
    meeting2_tmp = item.find('meeting2').text
    hwpLink_tmp = item.find('hwpLink').text
    summary_tmp = item.find('summary').text
    
    conferNum.append(conferNum_tmp)
    commName.append(commName_tmp)
    meeting1.append(meeting1_tmp)
    meeting2.append(meeting2_tmp)
    hwpLink.append(hwpLink_tmp)
    summary.append(summary_tmp)
    
# print(conferNum)
# print(commName)
# print(meeting1)
# print(meeting2)
# print(hwpLink)
# print(summary)
```

```python
# 데이터프레임 생성    
df = pd.DataFrame({'conferNum' : conferNum,
                   'commName' : commName,
                   'meeting1' : meeting1, 
                   'meeting2' : meeting2,
                   'hwpLink' : hwpLink,
                   'summary' : summary})

df
```

```python
df1 = df.sort_values('conferNum').reset_index(drop = True)
df1
```

```python
df1.rename(columns = {"conferNum":"confer_num"}, inplace = True)
df1.rename(columns = {"commName":"confer_name"}, inplace = True)
```
